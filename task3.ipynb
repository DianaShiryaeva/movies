Создание векторного представления

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_vectorizer.get_stop_words()
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(df['overview'])
tfidf_matrix.get_shape()
```

```python
def find_similar_movies(plot_description, top_n=5):
    user_vector = tfidf_vectorizer.transform([plot_description]) # Векторизация пользовательского описания

    cosine_similarities = cosine_similarity(user_vector, tfidf_matrix).flatten() # Вычисление косинусного сходства
    similar_indices = cosine_similarities.argsort()[-top_n:][::-1]

    results = df.iloc[similar_indices][['title', 'overview', 'imdb_rating', 'year',	'tagline', 'genres', 'production_countries']]
    return results
```

```python
user_plot = "A group of friends on an adventure to save the world."
similar_movies = find_similar_movies(user_plot)
```

```python
print("Похожие фильмы:\n")
for index, row in similar_movies.iterrows():
    print(f"Название: {row['title']}")
    print(f"Страна: {row['production_countries']}")
    print(f"Год: {int(row['year'])}")
    print(f"Рейтинг: {row['imdb_rating']}")
    print(f"Жанр: {row['genres']}")
    print(f"Краткое описание: {row['tagline']}")
    print(f"Описание: {row['overview']}\n")
```

```python
from transformers import AutoTokenizer, AutoModel
```

```python
model_ckpt = "sentence-transformers/multi-qa-mpnet-base-dot-v1"
model_ckpt
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
tokenizer
```

```python
model = AutoModel.from_pretrained(model_ckpt)
model
```

```python
import torch

device = torch.device("cuda")
model.to(device)
```

```python
from sentence_transformers import SentenceTransformer, util
```

```python
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]
```

```python
def get_embeddings(text_list):
    encoded_input = tokenizer(
        text_list, padding=True, truncation=True, return_tensors="pt"
    ).to(device)
    encoded_input = {k: v for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return cls_pooling(model_output)
```

```python
embedding = get_embeddings(df["overview"][0])
embedding.shape
```

```python
df["embeddings"] = df["overview"].map(lambda x: get_embeddings(x).detach().cpu().numpy()[0])
df
```

```python
df.to_csv("movies_with_embeddings.csv")
```

```python
import faiss
import numpy as np

index = faiss.IndexFlatL2(df["embeddings"][0].shape[0])
embeddings_np = np.array(df["embeddings"].tolist(), dtype=np.float32)
index.add(embeddings_np)
faiss.write_index(index, "my_faiss_index.bin")
```

```python
index = faiss.read_index("my_faiss_index.bin")

query = "A group of friends on an adventure to save the world"
query_embedding = get_embeddings([query])

k = 10
distances, indices = index.search(query_embedding.detach().cpu().numpy(), k)

print(f"Самые близкие соседи для '{query}'")
for i in range(k):
  print(f"{i+1}. Название: {df.iloc[indices[0][i]]['title']}")
  print(f"Страна: {df.iloc[indices[0][i]]['production_countries']}")
  print(f"Год: {df.iloc[indices[0][i]]['year']}")
  print(f"Рейтинг: {df.iloc[indices[0][i]]['imdb_rating']}")
  print(f"Жанр: {df.iloc[indices[0][i]]['genres']}")
  print(f"Краткое описание: {df.iloc[indices[0][i]]['tagline']}")
  print(f"Описание: {df.iloc[indices[0][i]]['overview']}")
```
